### Step 1: Prepare data. Turn it into a list of lists

```python
df = baskets
df.groupby('order_id').all()
```

```python
#####

transaction_data = []
for groups in df.groupby('order_id').groups.values():
    transaction_data.append(list(df.loc[groups]['sku_id'].to_numpy()))
```

```python
transaction_data[1080]
```

```python
def oneHotCoding(transaction_data):
    te = TransactionEncoder()
    te_ary = te.fit(transaction_data).transform(transaction_data)
    transaction_df = pd.DataFrame(te_ary, columns=te.columns_)
    return transaction_df

transaction_sku_data = oneHotCoding(transaction_data)
```

```python
transaction_sku_data
```

```python
def frq_sku_itemsets(data,suppVar):
    min_support = suppVar/data.shape[0]
    frequent_itemsets = fpgrowth(data, min_support=min_support, use_colnames=True)
    return frequent_itemsets

freq_itemsets = frq_sku_itemsets(transaction_sku_data,30)
```

```python
freq_itemsets
```

```python
rules = association_rules(freq_itemsets, metric="confidence", min_threshold=.97)
rules
```

```python
def basket_rules(freq_itemsets,metrics,threshold):
    asso_rules = association_rules(freq_itemsets, metric=metrics, min_threshold=threshold)
    return asso_rules.sort_values(by='lift', ascending=False)[['antecedents', 'consequents', 'support', 'confidence', 'lift']]

fp_rules = basket_rules(freq_itemsets,"confidence",0.99)
fp_rules.sort_values('support')
```

```python
## this function returns a list of relevent merchants who have the relevent consequents,
def consequent_list(consequent, merchants_with_antecedent):
    merchants_with_consequent = []
    for index in range(len(baskets_eval)):
        sku_id = baskets_eval.iloc[index,4]
        merchant_id = baskets_eval.iloc[index,3]
        if merchant_id in merchants_with_antecedent and sku_id == consequent and merchant_id not in merchants_with_consequent:
            merchants_with_consequent.append(merchant_id)    
    return merchants_with_consequent
    
## this function returns a list of all the relevent merchants who purchased each consequent
## in a given list
def multiple_consequent_check(consequent_array, list_of_antecedents):
    list_of_lists = []
    master_consequent_list = []
    length = len(consequent_array)
    for i in range(length):
        #print(i)
        list_of_lists.append(consequent_list(consequent_array[i], list_of_antecedents))
        #print(consequent_list(consequent_array[i], list_of_antecedents))
    #print (list_of_lists)
    list0_length = len(list_of_lists[0])
    for i in range(list0_length):
        checking_array = [1]
        for j in range(length-1):
            if (list_of_lists[0][i] in list_of_lists[j+1]):
                checking_array.append(j+1)
        if (len(checking_array) == length):
            master_consequent_list.append(list_of_lists[0][i])
    return master_consequent_list
```
